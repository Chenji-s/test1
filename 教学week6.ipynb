18.1. 元组 (Tuples)
元组与列表类似，但一个重要的区别是，元组是不可变的，也就是说，一旦定义了元组，不能修改其中的元素。元组使用小括号 () 来创建，而列表使用方括号 []。

```python
my_list = [4, 6, -2, 4, 0, 0]
my_tuple = (4, 6, -2, 4, 0, 0)

# 这是可以的 —— 列表是可变的
my_list[0] = 100
print(my_list)

# 这行代码会出错 —— 元组是不可变的
my_tuple[0] = 100
print(my_tuple)
```

18.2. 元组的索引和切片 (Indexing and Slicing)
与列表类似，元组也可以使用索引和切片操作来访问或提取其中的元素或子序列。

```python
a = (4, 6, -2, 4, 0, 0)  # 一个数字组成的元组
b = ()  # 一个空元组
c = (2, 2, (-4, 5), 2)  # 一个嵌套元组
d = (0.1, 'that', 2)  # 混合类型的元组

# 使用索引和切片
print(a[2:])   # 输出: (-2, 4, 0, 0)
print(c[2][0]) # 输出: -4
```

18.4. 解包 (Unpacking)
元组的一个有用特性是可以对元组进行解包。通过解包，我们可以将元组中的每个元素分别赋值给不同的变量。


```python
u, v, w = (3.4, 1, 'friday')
print(f'{u}, {w}')
#解释：上面的代码将元组中的元素依次赋值给变量 u、v 和 w。
#解包也可以用于交换两个变量的值：

# 交换两个值
u, w = w, u
print(f'{u}, {w}')
```

18.5. 函数的多重返回值 (Multiple Return Values)
当函数返回多个值时，实际上返回的是一个包含这些值的元组。通过解包，可以轻松将这些返回值赋值给不同的变量。



```python
def my_function(x):
    return x, x**2, 3*x - 1

# 解包返回的元组
a, b, c = my_function(10)
print(b) # 输出: 100
print(c) # 输出: 29
```

3. 推导式 (Comprehensions)
推导式是Python中简洁创建序列的方式。推导式可以用于列表、字典等类型，简化常见的 for 循环和条件语句。

3.1 列表推导式 (List Comprehensions)
列表推导式允许你使用一种紧凑的方式来创建新列表。基础语法是：

my_list = [expression for item in sequence if condition]

```python
square_numbers = [1, 4, 9, 16, 25]
square_plus1 = [x+1 for x in square_numbers]
print(square_plus1)  # 输出: [2, 5, 10, 17, 26]
```

```python
#等价
square_plus1 = []
for x in square_numbers:
    square_plus1.append(x+1)
print(square_plus1) 
```

3.2 其他推导式 (Other Comprehensions)

推导式不仅可以应用于列表，还可以用于字典。你可以用推导式来构建字典，通过 zip() 函数将两组列表元素配对成元组。

```python
cat_names = ['Olive', 'Muffin', 'Butterscotch', 'Mister Beans', 'Nala']
cat_ages = [7, 3, 12, 11, 5]

old_cats = {name: age for name, age in zip(cat_names, cat_ages) if age > 10}
print(old_cats)  # 输出: {'Butterscotch': 12, 'Mister Beans': 11}
```

3.3 嵌套推导式 (Nested Comprehensions)

推导式可以相互嵌套。例如，你可以用嵌套推导式生成小于100的质数。

```python
primes = [x for x in range(2, 100) if not [t for t in range(2, x) if not x % t]]
print(primes)
```

十九 递归函数（Recursive Functions）。

什么是递归？
递归是一种技术术语，表示一个过程或函数调用自身，以提供答案。这类似于数学中的递归关系。通过递归，可以用较少的代码解决一些复杂的问题。
递归函数的一般结构如下：

def recursive_function(x):
    # 终止条件
    if terminating_condition:
        return terminating_value
    else:
        # 递归条件
        return recursive_function(some_expression_depending_on_x)
        
Python实现阶乘的递归函数：

```python
def factorial(n):
    #计算n!，使用递归的方法。
    if n == 0:
        print('到底了... 返回 0! = 1')
        return 1
    else:
        print(f'正在返回 {n} * {n-1}!')
        return n * factorial(n-1)

print(factorial(5))
```

递归的工作原理
递归函数必须包含一个终止条件（terminating condition），否则会无限循环。阶乘函数的终止条件是 n == 0，这时返回1，然后返回给上一级调用。否则，递归函数会调用自身，直到遇到终止条件。

例如，输入 3 时，调用过程如下：
3×fac(2)×fac(1)×fac(0)
然后结果会逐级返回，从最底层 fac(0) 返回到 fac(3)。

练习：编写一个递归函数 fib_rec(p, q, n) 来计算斐波那契数列的第 n 项，定义如下：
F(1)=F(2)=1,F(n)=pF(n−1)+qF(n−2),for n>2
测试：
fib_rec(1, 1, 15) 应返回 610
fib_rec(6, 4, 10) 应返回 5330944

```python
def fib_rec(p, q, n):
    '''
返回(p,q)-Fibonacci序列的第n个元素。
输入:
P (int)，正整数，第一系数
Q (int)，正整数，第二系数
N (int)，正整数
输出:
Fn (int)，序列的第n个元素，定义为
F(n) = pF(n-1) + qF（n-2）
    '''
    if n == 1 or n == 2:
        return 1
    else:
        return p*fib_rec(p, q, n-1) + q*fib_rec(p, q, n-2)

assert fib_rec(1, 1, 15) == 610
assert fib_rec(6, 4, 10) == 5330944
assert fib_rec(2, 1, 35) == 5168247530883
```

2. 解决问题的示例：邮局问题
这个问题是一个最近邻查找问题，其描述如下：

“假设在一个5公里乘5公里的方形城市中，分布着n个邮局，每个邮局的位置是已知的。你需要生成一张地图，帮助居民找到距离他们最近的邮局。”

在这个问题中，你需要设计一个算法帮助解决居民与最近邮局的距离问题。为了实现这个目标，代码示例创建了n个邮局，随机散布在城市内。

```python

import numpy as np

def create_POs(n):
    '''
    创建城市中的n个邮局的随机坐标。
    
    输入：n（整数），邮局总数。
    输出：POs（ndarray），大小为n x 2的随机数组，每行给出一个邮局的(x, y)坐标。
    '''
    rng = np.random.default_rng()
    POs = 0.2 + 4.6 * rng.random(size=[n, 2])  # 生成坐标
    return POs
```

让我们来看看这些邮局，以便更好地了解我们在做什么。将数据或结果可视化总是一个好主意。

```python
import matplotlib.pyplot as plt

# Create 10 post offices
n = 10
POs = create_POs(n)

# Plot their location
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(POs[:, 0], POs[:, 1], 'k*', markersize=15)
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
plt.show()
```

2.1 设计算法
为了设计这个算法，作者建议通过以下检查清单来帮助你分解问题并规划解决方案：

问题的最终目标是什么？
需要生成城市的地图。
地图应该帮助居民选择最近的邮局。
可以将城市分为不同区域，表示每个邮局的“服务范围”。
我有什么数据？
我们知道邮局的位置（它们的(x, y)坐标）。
我们知道城市的大小（5 x 5公里方形区域）。
我们可以通过曼哈顿距离来计算每个点之间的距离。
算法的粗略结构是什么？
生成随机居民坐标，找到与他们最近的邮局，使用np.linalg.norm()函数来计算曼哈顿距离。
使用np.argmin()找到最小的距离，标识每个居民最近的邮局。

为了逐步解决问题，代码通过以下步骤来模拟城市中的居民，并计算他们到各个邮局的距离：

```python
def make_residents(pop):
    '''
    创建城市中均匀分布的居民人口。
    
    输入：pop（整数），人口数量。
    输出：居民的坐标（ndarray），大小为pop x 2的数组。
    '''
    rng = np.random.default_rng()
    residents = 5 * rng.random(size=[pop, 2])  # 居民坐标
    return residents

def nearest_PO(residents, POs):
    '''
    为所有居民找到最近的邮局。
    
    输入：
        residents（ndarray）：居民的(x, y)坐标数组。
        POs（ndarray）：邮局的(x, y)坐标数组。
    输出：
        最近邮局的索引（ndarray）：对应每个居民的最近邮局的索引。
    '''
    distances = []
    
    for po in POs:
        dist_po = []
        for res in residents:
            dist_po.append(np.linalg.norm(res - po, ord=1))  # 计算曼哈顿距离
        distances.append(dist_po)
    
    distances = np.array(distances).T  # 转换为NumPy数组
    closest = np.argmin(distances, axis=1)  # 找到最近的邮局
    return closest

# Run this cell a few times to test different random values
residents = make_residents(8)
POs = create_POs(3)
closest = nearest_PO(residents, POs)
print(closest)
```

这一切看起来都很合理——当然，我们不知道结果是否正确。现在，让我们创建映射绘制函数，因为这将为我们提供一种将结果可视化并正确检查它们的方法。

我们可以用相同的颜色画出每个居民，如果他们有相同的最近的邮局。我们可以使用布尔索引来决定绘制特定颜色的居民-这将比用大循环单独绘制每个点快得多！

我们将循环遍历每个邮局，并使用该邮局的一个命令绘制所有居民。我们需要为每个PO选择不同的颜色——我们可以提前选择颜色，并创建一个字符串列表，以便在调用.plot（）时使用它。或者，我们可以从matplotlib中选择一个colormap对象，使用plt.get_cmap（）检索它，然后在循环遍历邮局时，在渐变的某个地方选择一种颜色。

```python
def draw_PO_map(residents, POs):
    '''
    显示居民分布在地图上，并标记他们最近的邮局，用不同的颜色区分。
    
    输入:
        residents (ndarray): 一个具有 "pop" 行和 2 列的数组，
            每一行表示一位居民的 (x, y) 坐标。
        POs (ndarray): 一个具有 n 行和 2 列的数组，每一行
            表示一个邮局的 (x, y) 坐标。
    
    输出: 无
    '''
    # 获取居民数量和邮局数量
    pop = residents.shape[0]
    n = POs.shape[0]
    
    # 设置绘图框架
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect('equal')
    ax.set_xlim([0, 5])
    ax.set_ylim([0, 5])
    
    # 获取邮局的颜色图
    colour_map = plt.get_cmap('gist_rainbow')
    colours = []
    for c in range(n):
        # 提取一种颜色，位于颜色图中的某个位置。
        # colour_map(0) 返回 'gist_rainbow' 图中的最左边的颜色，
        # colour_map(1) 返回最右边的颜色。
        colours.append(colour_map(c/n))

    # 使用之前创建的函数找到每位居民最近的邮局
    closest = nearest_PO(residents, POs)
    
    # 用不同颜色绘制每个居民集群
    for po in range(n):
        # 布尔索引：提取 "closest" 为当前邮局的居民行
        x = residents[closest == po, 0]
        y = residents[closest == po, 1]
        ax.plot(x, y, '.', color=colours[po], markersize=8)
    
    # 绘制每个邮局的标记
    ax.plot(POs[:, 0], POs[:, 1], 'k*', markersize=15)
    plt.show()

# 决定我们想要多少居民，并创建他们
pop = 10000
residents = make_residents(pop)

# 在城市中创建邮局
n = 30
POs = create_POs(n)

# 现在，绘制地图！
draw_PO_map(residents, POs)
```

这个练习的目的是通过模拟居民和邮局的分布，帮助我们了解邮局布局对于城市服务效率的影响。

```python
# 我们可以重新定义我们的 nearest_PO 函数，使其同时返回距离
def nearest_PO(residents, POs, metric):
    '''
    找到每位居民最近的邮局。
    
    输入:
        residents (ndarray): 一个包含 "pop" 行和 2 列的数组，
            每一行是一个居民的 (x, y) 坐标。
        POs (ndarray): 一个包含 2 列的数组，每一行是一个
            邮局的 (x, y) 坐标。
        metric (str): 'manhattan' 表示曼哈顿距离，'euclid' 表示欧几里得距离。
    
    输出:
        closest (ndarray): 每位居民最近邮局的索引，
            基于欧几里得距离。
    '''
    if metric == 'manhattan':
        order = 1
    elif metric == 'euclid':
        order = 2
    else:
        print('请输入一个有效的度量标准！')
        return
    
    # 准备一个列表来存储所有的距离
    distances = []
    
    # 遍历邮局
    for po in POs:
        dist_po = []
        # 遍历居民
        for res in residents:
            # 计算每位居民与邮局之间的 2-范数距离
            dist_po.append(np.linalg.norm(res - po, ord=order))
        
        # 将所有居民到该邮局的距离列表加入 distances 中
        distances.append(dist_po)
    
    # 将列表转换为 NumPy 数组
    distances = np.array(distances).T
    
    # 返回每位居民最近邮局的索引（找到每位居民的最小距离）
    closest = np.argmin(distances, axis=1)
    
    # 保存最短的距离
    pop = residents.shape[0]
    dist_closest = np.zeros(pop)
    for i in range(pop):
        dist_closest[i] = distances[i, closest[i]]
    
    return closest, dist_closest


def average_time(speed, pop):
    '''
    绘制平均步行时间（到最近邮局）与邮局数量之间的函数关系。
    
    输入:
        speed (float): 步行速度，单位：km/h
        pop (int): 模拟中的人口数量。
    
    输出: 无
    '''
    av_time = np.zeros([30, 2])
    residents = make_residents(pop)
    
    # 遍历邮局数量
    for n in range(1, 31):
        POs = create_POs(n)

        # 使用两种度量标准获取到最近邮局的距离
        _, dist1 = nearest_PO(residents, POs, 'manhattan')
        _, dist2 = nearest_PO(residents, POs, 'euclid')

        # 计算平均距离和时间
        av_dist1 = np.mean(dist1)
        av_time[n-1, 0] = av_dist1 / speed
        av_dist2 = np.mean(dist2)
        av_time[n-1, 1] = av_dist2 / speed
    
    # 绘制结果
    fig, ax = plt.subplots()
    ax.plot(range(1, 31), av_time[:, 0], '-', label='曼哈顿距离')
    ax.plot(range(1, 31), av_time[:, 1], '-', label='欧几里得距离')
    ax.legend()
    ax.set_xlabel('邮局数量')
    ax.set_ylabel('平均步行时间（小时）')
    plt.show()

# 这部分运行需要一点时间 -- 可以从较小的人口数量开始
average_time(5, 1000)
```
