1. 轴范围（Axis limits）
可以使用 .set_xlim() 和 .set_ylim() 方法控制图中的轴范围。例如，以下代码绘制函数
sin(x) 在区间 x∈[0,2π] 上的图像，并将 x 轴的范围设置为 0 到 2π，y 轴的范围为 -1.1 到 1.1：

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建 x 轴的 100 个值，范围从 0 到 2π
x = np.linspace(0, 2 * np.pi, 100)

# 创建图像和坐标轴
my_fig, my_ax = plt.subplots()

# 绘制 sin(x)
my_ax.plot(x, np.sin(x), 'k-')

# 调整 x 和 y 轴的范围
my_ax.set_xlim([0, 2 * np.pi])
my_ax.set_ylim([-1.1, 1.1])

#2. 轴标签（Axis labels）
#可以使用 .set_xlabel() 和 .set_ylabel() 方法添加轴标签。例如：
my_ax.set_xlabel('x', fontsize=12)
my_ax.set_ylabel(r'$y$', fontsize=12)

# 显示图像
plt.show()
```


练习 15.1
复制第十四章第2题的代码，将三个函数分别绘制在三个子图中。使用 set_xlim 控制每个图的 x 轴范围，使其在 x∈[−π,π] 之间。使用 set_xlabel 和 set_ylabel 为每个子图添加标签，字体大小为 14。

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x
x = np.linspace(0, 2*np.pi, 1000)

# Create figure and axes
fig_1, ax_1 = plt.subplots(1, 3,figsize=(10, 4))

ax_1[0].plot(x, np.sin(x), 'r-.')
ax_1[1].plot(x, np.cos(x), 'g:')
ax_1[2].plot(x, np.cos(x), 'b--')

# Store y-axis label for each plot
y_labels = [r'$f_1(x)$', r'$f_2(x)$', r'$f_3(x)$']

# Set all 3 properties for the 3 plots
for i in range(3):
    ax_1[i].set_xlim([-np.pi, np.pi])
    ax_1[i].set_xlabel(r'$x$', fontsize=14)
    ax_1[i].set_ylabel(y_labels[i], fontsize=14)

# Make some space
plt.subplots_adjust(hspace=0.5, wspace=0.5)
# Show the result
plt.show()
```
