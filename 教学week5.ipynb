1. 轴范围（Axis limits）
可以使用 .set_xlim() 和 .set_ylim() 方法控制图中的轴范围。例如，以下代码绘制函数
sin(x) 在区间 x∈[0,2π] 上的图像，并将 x 轴的范围设置为 0 到 2π，y 轴的范围为 -1.1 到 1.1：

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建 x 轴的 100 个值，范围从 0 到 2π
x = np.linspace(0, 2 * np.pi, 100)

# 创建图像和坐标轴
my_fig, my_ax = plt.subplots()

# 绘制 sin(x)
my_ax.plot(x, np.sin(x), 'k-')

# 调整 x 和 y 轴的范围
my_ax.set_xlim([0, 2 * np.pi])
my_ax.set_ylim([-1.1, 1.1])

#2. 轴标签（Axis labels）
#可以使用 .set_xlabel() 和 .set_ylabel() 方法添加轴标签。例如：
my_ax.set_xlabel('x', fontsize=12)
my_ax.set_ylabel(r'$y$', fontsize=12)

# 显示图像
plt.show()
```

题目要求：

使用一个坐标轴绘制函数 cos(x) 和它的 Maclaurin 展开式。
在同一个坐标轴上绘制 Maclaurin 展开式的二阶、四阶和六阶近似。
设置坐标轴范围、标签，并为绘图添加图例。
提示：

可以使用 math.factorial() 函数来计算阶乘。

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# 为截断的Maclaurin级数定义一个函数
def trunc_cos(x, n):
     #返回为的截断的麦克劳林系列 Cos (x)直到(n)
    cos_series = 0
    for k in range(n//2 + 1):
        # 把级数的每一项加到n阶
        cos_series += (-1)**k * x**(2*k) / math.factorial(2*k)
    
    return cos_series


# Create an x-axis with 1000 points
x = np.linspace(-np.pi, np.pi, 1000)

# Create a figure
fig, ax = plt.subplots()

# Plot the requested functions
ax.plot(x, np.cos(x), 'k-', label=r'$\cos(x)$')
ax.plot(x, trunc_cos(x, 2), 'r--', label=r'Order 2')
ax.plot(x, trunc_cos(x, 4), 'g-.', label=r'Order 4')
ax.plot(x, trunc_cos(x, 6), 'b:', label=r'Order 6')

# Set axis properties
ax.set_xlim([-np.pi, np.pi])
ax.set_ylim([-1.5, 1.5])
ax.set_xlabel(r'$x$', fontsize=12)
ax.legend()

plt.show()
```

第十六章 Docstrings
Docstring 是一种写在Python函数定义内部的注释，用于描述函数的功能、输入参数以及输出结果。它们位于 def 语句下方，用 ''' 三重引号包含。这些注释的目的是记录函数的作用，帮助用户理解如何使用函数及其返回值。

通常，一个docstring应该包含以下信息：

一两句话概括函数的功能；
描述所有输入参数及其类型；
描述所有输出及其类型。
这种做法确保使用该函数的用户能够知道函数的用途、作用对象及返回内容。

```python
def first_and_last(x):
    '''
    返回字符串的首尾字符。

    输入：
    x (str): 输入字符串。

    输出：
    out (str): 一个包含输入字符串首尾字符的新字符串。
    '''
    out = x[0] + x[-1]
    return out
x=[1,2,3,4]
first_and_last(x)
```

函数 random_letters 从输入字符串 word 中随机抽取 number_of_letters 个字母，并将这些字母存入一个列表。
random.randint(0, len(word)-1) 生成一个在 0 到 len(word)-1 范围内的随机索引，用于从 word 中随机选择字母。

```python
import random

def random_letters(word, number_of_letters):
    list_of_letters = []

    for i in range(number_of_letters):
        index = random.randint(0, len(word)-1)
        letter = word[index]
        list_of_letters.append(letter)

    return list_of_letters
w=['re','ee','ii']
random_letters(w,2)
```

第十七章 SciPy库
SciPy 是 Python 的科学计算库，它与 NumPy 有着紧密的联系，主要用于数值计算。SciPy 包含许多模块，可以帮助解决各种实际问题。以下是一些重要的模块：

Optimization：用于求解函数的最值以及方程的根。
Interpolation：用于插值（例如样条插值）。
Statistics：提供统计相关的函数。
Integration：用于数值积分。

例子：插值噪声数据 (Example: interpolating some noisy data)
假设我们有一组数据点 （xi，yi），数据存储在向量 x 和 y 中。我们可以使用 scipy.interpolate 模块来进行数据插值，尤其是使用样条插值。

以下是一个样条插值的例子。

```python
# 导入必要的模块
import numpy as np
import scipy.interpolate
import scipy as scp
import matplotlib.pyplot as plt

# 定义生成噪声数据的函数
def generate_data(x):
    '''
    生成一些带噪声的平滑数据 (正弦曲线)。
    '''
    # 从正态分布生成噪声
    noise = np.random.normal(0, 0.2, len(x))
    y = np.sin(x) + noise
    return y

# 设置 x 并生成 y 数据
step = 0.4
x = np.arange(0, 10 + step, step)
y = generate_data(x)

# 使用分段三次函数进行插值
f_interp = scp.interpolate.interp1d(x, y, kind='cubic')

# 使用另一种方法进行光滑插值
f_smooth = scp.interpolate.UnivariateSpline(x, y, s=2)

# 假设我们希望在 x=1.5 时求 y 的值
print(f_interp(1.5))
print(f_smooth(1.5))
```

```python
# 绘制插值结果
# 设置密集的 x 轴用于绘制插值结果
xmin, xmax = x[0], x[-1]
dense_step = 0.05
x_dense = np.arange(xmin, xmax + dense_step, dense_step)

# 绘制数据
fig, ax = plt.subplots()
ax.plot(x, y, 'k.', label='Noisy data')
ax.plot(x_dense, np.sin(x_dense), 'g-', label='Underlying function')
ax.plot(x_dense, f_interp(x_dense), 'c--', label='Interpolating spline')
ax.plot(x_dense, f_smooth(x_dense), 'r--', label='Smooth spline')

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_xlim([xmin, xmax])
ax.legend(loc='upper center')

plt.show()
```

2. 示例：汇总统计信息，读取文件 (Example: summary statistics, reading a file)
这一部分介绍了如何读取包含能源生产百分比数据的CSV文件 renewable_pc.csv，并使用Numpy和SciPy库进行基本的汇总统计。


np.loadtxt()：这是一个简单的函数，可以用于加载文本文件中的数据。参数解释：
dtype=float：指定数据类型为浮点数。
delimiter=','：指定文件是逗号分隔的CSV文件。
skiprows=1：跳过文件中的第一行（通常是标题行）。


```python
import numpy as np
import scipy.stats as stats

# 使用Numpy的loadtxt函数打开数据文件：
energy_data = np.loadtxt('renewable_pc.csv', dtype=float, delimiter=',', skiprows=1)
#将每个列解压缩到一个不同的数组中
#数据中描述的国家和年份（使用转置到）
#获取数组的行列表)：
year, bulgaria, croatia, denmark, iceland = energy_data.T

#使用description函数获取基本的汇总统计信息
summary_stats = stats.describe(croatia)
print(summary_stats)

#查找第一和第三个四分位数：
print('First and third quartiles:')
print(np.percentile(iceland, (25, 75)))

#找出皮尔逊和斯皮尔曼的相关系数,在两列之间：
print('Correlation coefficients for Croatia/Bulgaria and Croatia/Iceland:')
print(stats.pearsonr(croatia, bulgaria))
print(stats.spearmanr(croatia, iceland))

#我们可以使用Numpy's将数组保存到文件中,savetext（）函数：
np.savetxt('denmark.txt', denmark)
```

在本练习中，要求你完成以下任务：

计算不同列之间的相关性。
查找丹麦和冰岛数据之间的线性关系。


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib as mpl

# 保加利亚统计数字
mode, count = stats.mode(bulgaria)
print('Bulgaria:\n',
      f'Median = {np.median(bulgaria):.1f}\n',
      f'Geometric mean = {stats.gmean(bulgaria):.2f}\n')


# 计算所有国家对之间的皮尔森相关系数
countries = ['Bulgaria', 'Croatia', 'Denmark', 'Iceland']

pearson_coeffs = []

# 忽略第一个 "Year" 列，计算国家间的相关系数
for country1 in energy_data.T[1:]:
    country1_coeffs = []
    for country2 in energy_data.T[1:]:
        country1_coeffs.append(np.round(stats.pearsonr(country1, country2)[0], 2))
    
    pearson_coeffs.append(country1_coeffs)

# 使用颜色映射显示相关系数表
colourmap = mpl.colormaps['Greens']
colours = colourmap(pearson_coeffs)

fig, ax = plt.subplots(1, 1, figsize=(7, 2))
ax.table(cellText=pearson_coeffs,
         rowLabels=countries,
         colLabels=countries,
         cellColours=colours,
         loc='center',
         cellLoc='center')
ax.axis('tight')
ax.axis('off')
plt.show()

# 绘制保加利亚与克罗地亚的数据点
fig, ax = plt.subplots(1, 1, figsize=(7, 5))
ax.plot(bulgaria, croatia, 'kx', label='Data points')
ax.set_xlabel('Renewable share for Bulgaria')
ax.set_ylabel('Renewable share for Croatia')

# 使用线性回归模型拟合数据
regr = stats.linregress(bulgaria, croatia)
print(regr)

# 绘制回归线
x_values = np.linspace(bulgaria.min(), bulgaria.max(), 100)
y_values = x_values * regr.slope + regr.intercept
ax.plot(x_values, y_values, 'r-', label=f'Linear fit (R^2 = {regr.rvalue**2:.2f})')

ax.legend()
plt.show()
```

17.3. 示例：优化 (Example: optimisation)
这一部分介绍了使用 scipy.optimize 库进行线性规划问题求解的基本方法。问题描述的是一名木匠需要在工作时间内制作桌子和椅子，目标是最大化他的利润。每张桌子能卖30英镑，每把椅子能卖10英镑，制作桌子需要6小时，制作椅子需要3小时，木匠每周最多有40个小时的工作时间。此外，顾客要求椅子数量至少是桌子的3倍，桌子占用的存储空间是椅子的4倍，木匠每周最多只能存放4张桌子。

```python
from scipy.optimize import linprog

A = np.array([[6, 3], [3, -1], [1, 0.25]])
b = np.array([40, 0, 4], dtype=float)
c = np.array([-30, -10], dtype=float)

solution = linprog(c, A_ub=A, b_ub=b)
print(solution)
```
